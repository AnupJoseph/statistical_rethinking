Let ignore maths for a moment and calculate the numbers in reality
* For each possible value of *p*
* Compute product $$Pr(W,L|p)Pr(p)$$ 
* Relative sizes of the products in (2) are posterior probabilities

```python
def build_posterior(num_steps=1000, num_tries=4, total_tries=9):
    p_grid = np.arange(0, 1, step=1 / num_steps)
    prob_p = np.repeat(1, num_steps)
    prob_data = binom.pmf(num_tries, total_tries, p=p_grid)
    posterior = prob_data * prob_p
    posterior = posterior / sum(posterior)
    return posterior
```
